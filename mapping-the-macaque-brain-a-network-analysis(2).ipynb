# Introduction:

In this project, we will be focusing on analyzing the brain areas and connections of the macaque brain. We will use the dataset **macaque** from **igraphdata** package.  This dataset is made available for those who are interested in neuroscience and its applications in real life. The macaque brain is often used as a model system for the human brain due to its similarities to humans.

<h4>Libraries and Dataset:</h4>

We will be using the **igraph and ggraph** library to analyze the macaque brain dataset. Let's begin by installing and loading the igraph library.

```r
library(igraph)
library(igraphdata)
library(ggraph)
```

```r
data("macaque")
```

<h4>Degree Centrality:</h4>

In graph theory and network analysis, degree centrality is the number of edges connected to a node. The more edges that are connected to a node, the more central it is to the network.

To calculate the degree centrality of each node in the macaque dataset, we can use the **degree()** function in igraph. We store the results in a variable called **macaque_degree_centrality**.

```r
macaque_degree_centrality <- degree(macaque)
macaque_degree_centrality
```

This outputs the degree centrality of each node in the macaque dataset. Each node is identified by a letter and a number (e.g., V1, V2, etc.). Higher values indicating more connections to other nodes.

<h4>Betweenness Centrality:</h4>

Betweenness centrality measures the extent to which a node lies on paths between other nodes. A node with high betweenness centrality has a large influence on the transfer of information in a network, because a large proportion of information will pass through it.

To calculate the betweenness centrality of each node in the macaque dataset, we can use the **betweenness()** function in igraph. We store the results in a variable called **macaque_betweenness_centrality**.

```r
macaque_betweenness_centrality <- betweenness(macaque)
macaque_betweenness_centrality
```

This outputs the betweenness centrality of each node in the macaque dataset. Higher values indicating that the node lies on more shortest paths between other nodes.

<h4>Closeness Centrality:</h4>

Closeness centrality is a measure of the average distance from a node to all other nodes in the network. It is based on the idea that a node that is close to all other nodes in the network is more central than a node that is far away from other nodes.

To calculate the closeness centrality of each node in the macaque dataset, we can use the **closeness()** function in igraph. We store the results in a variable called **macaque_closeness_centrality.**

```r
macaque_closeness_centrality <- closeness(macaque)
macaque_closeness_centrality
```

This outputs the closeness centrality of each node in the macaque dataset. Higher values indicating that the node is closer to all other nodes.

<h4> Visualisation</h4>

We can now use ggraph to visualize the macaque network, highlighting the degree, betweenness, and closeness centrality of each node.

```r
ggraph(macaque, layout = "fr") +
  geom_edge_link() +
  geom_node_point(aes(size = macaque_degree_centrality))
```

The first graph shows the macaque network with the size of each node representing its degree centrality. In this visualization, we can see the nodes with the most edges connected to them are larger, indicating they have higher degree centrality. We can also see how these highly connected nodes are distributed throughout the network, and identify any potential clusters or sub-networks within the macaque brain.

```r
ggraph(macaque, layout = "fr") +
  geom_edge_link() +
  geom_node_point(aes(size = macaque_betweenness_centrality))
```

The second graph shows the macaque network with the size of each node representing its betweenness centrality. In this visualization, we can see the nodes that are the most central to the network, meaning they lie on the shortest path between other nodes, are larger in size. These central nodes may play important roles in information processing and communication within the macaque brain.

```r
ggraph(macaque, layout = "fr") +
  geom_edge_link() +
  geom_node_point(aes(size = macaque_closeness_centrality))
```

The third graph shows the macaque network with the size of each node representing its closeness centrality. In this visualization, we can see the nodes that are the most easily accessible to other nodes in the network are larger in size. These nodes may be important for quick and efficient transmission of information within the macaque brain.

We can also identify the top 10 nodes by degree centrality and visualize their relationships with other nodes in the macaque network.

```r
degree <- degree(macaque)

top_nodes <- names(sort(degree, decreasing = TRUE))[1:10]

top_nodes
```
